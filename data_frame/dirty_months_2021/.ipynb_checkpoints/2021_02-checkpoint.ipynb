## Clean the Data

```python
#Change dtypes if necessary
```

```python
#Remove "Status" column from df_2021_02.
df_2021_02 = df_2021_02.drop('Status', axis=1)
df_2021_02
```

```python
#Remove "Zone" column from df_2021_02.
df_2021_02 = df_2021_02.drop('Zone', axis=1)
df_2021_02
```

```python
# Drop the NaNs for df_2021_02.
df_2021_02 = df_2021_02.dropna()
```
