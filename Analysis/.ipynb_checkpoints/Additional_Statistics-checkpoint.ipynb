```python
#Pandas library to create DataFrames
import pandas as pd
#NumPy library for converting data types
import numpy as np
```

```python
winedemic_df = pd.read_csv("../winedemic_complete.csv.zip")
```

```python
winedemic_df.head()
```

## 2019

```python
year2019_df = winedemic_df[winedemic_df["Year"] == 2019]
year2019_df.head(5)
```

```python
year2019_df = year2019_df.set_index("State")
year2019_df.head(5)
```

```python
# Calculate the total 2019 Item/Bottle Count.
total_count_2019 = year2019_df["Item/Bottle Count"].sum()
total_count_2019
```

```python
# Use .groupby() to find item/bottle count .sum() for each state
total_bottles_by_state_2019 = year2019_df.groupby(["State"]).sum()
total_bottles_by_state_2019.head(5)
```

```python
total_bottles_by_state_2019 = total_bottles_by_state_2019.drop(columns=["Weight", "Year", "Day",
                                                                        "Month", "Unnamed: 0"], axis=1)
```

```python
total_bottles_by_state_2019.sort_values(by='Item/Bottle Count', ascending=False)
```
