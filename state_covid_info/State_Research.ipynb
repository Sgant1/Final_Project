## Amanda Cleaning

```python
#Remove "Unnamed:15" and "Unnamed: 14" column from working_state_Amanda.
working_state_Amanda = working_state_Amanda.drop('Unnamed: 15', axis=1)
```

```python
working_state_Amanda = working_state_Amanda.drop('Unnamed: 14', axis=1)
working_state_Amanda
```

```python
#Create df that has "State", "Population", "Lockdown Start", "Initial Expected Lockdown End", "Phase 1 Re-Opening", "Max Shipping Amount", "Politial Alignment Red vs. Blue 2020"
new_amanda_df = working_state_Amanda.filter(["State", "Population", "Lockdown Start", "Inital Expected\nLockdown End", "Phase 1\nRe-Opening Start", "Politcal Alignment Red vs. Blue 2020"])
new_amanda_df
```

```python
new_amanda_df.rename(columns={'Inital Expected\nLockdown End': 'Inital Expected Lockdown End', 
                              'Phase 1\nRe-Opening Start': 'Phase 1 Re-Opening Start', 
                             'Politcal Alignment Red vs. Blue 2020': 'Political Alignment'}, inplace=True)
new_amanda_df
```

```python
#complete_states_df['Inital Expected Lockdown End'] = pd.to_datetime(complete_states_df,'Inital Expected Lockdown End')
new_amanda_df["Inital Expected Lockdown End"] = new_amanda_df["Inital Expected Lockdown End"].dt.strftime('%m/%d/%Y')
```

```python
#Want to change the Nan to a generic 00/00/0000 date or note somehow that there is no information.
new_amanda_df["Inital Expected Lockdown End"] = new_amanda_df["Inital Expected Lockdown End"].fillna("00/00/000"), inplace=True
new_amanda_df
```
